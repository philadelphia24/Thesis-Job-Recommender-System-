## Removing jobseekers with <three ratings

For better model performance, i.e. removing noise.

Let's remove the 3605 jobseekers who have rated less than three, so either one or two, vacancies.

```python
# Count the number of jobseekers who have rated less than 5 vacancies
jobseekers_with_few_ratings = matches.groupby('JobseekerID')['Rating'].count()
jobseekers_with_few_ratings = jobseekers_with_few_ratings[jobseekers_with_few_ratings < 3].index
num_jobseekers_with_few_ratings = len(jobseekers_with_few_ratings)

print(f"There are {num_jobseekers_with_few_ratings} jobseekers who have rated less than three, so either one or two, vacancies.")
```

```python
# Filter out jobseekers with less than three rated vacancies
matches_shortened = matches[~matches['JobseekerID'].isin(jobseekers_with_few_ratings)]
```

```python
display(matches_shortened)
```

```python
matches_shortened = matches_shortened.drop(['Name', 'Language', 'Education Level', 'Industry', 'Experience'], axis=1)
```

```python
display(matches_shortened)
```
